```python
"""we will make list of all the null values present in the dataset"""
```

```python
fea=[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
fea
```

```python
for i in fea:
    print(i,np.round(df[i].isnull().mean(),4),"% missing values")
```

```python
"""we should make some of the relationship between dependent and independent features"""
```

```python
 data= df.copy()
for feature in fea:
  #data= df.copy()
  data[feature]=np.where(data[feature].isnull(),0,1)#replacing nan with something meaningful 
  data.groupby(feature)["SalePrice"].median().plot.bar()#gropuby function actually takes the other columns data and means it
  plt.title(feature) 
  plt.show()#if we donot use it it wil show only last part of the loop 
```
